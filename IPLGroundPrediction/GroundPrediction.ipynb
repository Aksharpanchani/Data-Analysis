```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('./ball_by_ball_ipl.csv')

def find_probability(run_threshold):
    # Filtering data (I filtered this data according to the question we had to solve)
    # That is I would check for 2nd Innings and death overs are from 18 and this is grouped by according to the venue.
    
    chasing_data = data[data['Innings'] == 2]
    chasing_data['In Last 3 Overs'] = chasing_data['Over'] >= 18
    death_overs = chasing_data[chasing_data['In Last 3 Overs']].groupby(['Match ID', 'Venue'])['Runs From Ball'].sum().reset_index()
    
    # Merging data for the original outcomes.
    match_result = data[['Match ID', 'Venue', 'Chased Successfully']].drop_duplicates()
    result_data = pd.merge(death_overs, match_result, on=['Match ID', 'Venue'])
    
    # Calculate for the chasing team who won the match
    result_data[f'Successful {run_threshold}+ Runs'] = (result_data['Runs From Ball'] >= run_threshold) & (result_data['Chased Successfully'] == 1)
    
    # Grouping by venue to calculate the total matches and successful chases
    Total_venue = result_data.groupby('Venue').agg(
        Total_Matches=('Match ID', 'size'),
        Successful_Chases=(f'Successful {run_threshold}+ Runs', 'sum')
    ).reset_index()
    
    # Calculating the probability 
    Total_venue['Probability'] = Total_venue['Successful_Chases'] / Total_venue['Total_Matches']
    
    # Sorting the values
    Total_venue = Total_venue.sort_values(by='Probability', ascending=False)
   
    #Printing Probability
    #We are checking this in order to cross check whether there is any venue or not where the team has won.
    #This helps us to prove that there is no grounds where >60 are chased according to the dataset used.
    #For runs other than 60 the values would come as true (Can be seen for 50, 40, 30 runs....)
    #There are 843 entries which are filtered out
    print(result_data)

    # Plot the results
    plt.figure(figsize=(10, 3))
    plt.bar(Total_venue['Venue'], Total_venue['Probability'])
    plt.xticks(rotation=90)
    plt.ylim(0, 1)
    plt.xlabel('Stadium')
    plt.ylabel('Probability')
    plt.title(f'Probability for {run_threshold} to Win the Game')
    plt.show()

# Plot for different thresholds
find_probability(60)
find_probability(50)
find_probability(40)
find_probability(30)
find_probability(20)
```
